#### Item D

Para reduzir dimensionalidade e ter uma melhor visualização, foi utilizado o PCA (Principal Component Analysis), que é um procedimento estatístico que usa transformações ortogonais para converter um set de variáveis correlacionaveis. PCA seria melhor utilizado para classificação, do que para visualização.

```python
x = df.loc[:, boston.feature_names].values

y = df.loc[:,['CHAS']].values

dfscalled = StandardScaler().fit_transform(x)

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(dfscalled)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

pcaplot=plt.scatter(principalDf['principal component 1'],principalDf['principal component 2'])
```
